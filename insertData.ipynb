```python
import pandas as pd

def check_dataframe(df):
    """
    Perform various checks on the DataFrame.
    
    Parameters:
    - df (pd.DataFrame): The DataFrame to check.
    
    Returns:
    - None
    """
    # Check for missing values
    print("Missing values in each column:")
    print(df.isnull().sum())
    print("\n")

    # Check for duplicate rows
    duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicates}")
    print("\n")

    # Check data types of each column
    print("Data types of each column:")
    print(df.dtypes)
    print("\n")

    # Check unique values for specific columns (customize as needed)
    for column in df.columns:
        unique_values = df[column].unique()
        print(f"Unique values in column '{column}':")
        print(unique_values[:10])  # Display only the first 10 unique values for brevity
        print("\n")

    # Check for unexpected values in 'price' column (example)
    if 'price' in df.columns:
        unexpected_prices = df[~df['price'].astype(str).str.isnumeric()]
        print("Rows with unexpected price values:")
        print(unexpected_prices)
        print("\n")

    # Add more checks as needed

# Example usage
# df = pd.read_csv('your_file.csv')  # Load your DataFrame
# check_dataframe(df)
```

```python
# insert Product

directory_path = "json/product"
category_names = os.listdir(directory_path)
aa = []
print(category_names)
for i in category_names:
    product_directory = "json/product/" + i + "/"
    path_product_json = product_directory + os.listdir(product_directory)[0]
    product_df = pd.read_json(path_product_json).get(i)
    for row in product_df:
        aa.append(row)
Data = [i for i in aa]
Product = pd.DataFrame(Data).reset_index()
Product.rename(columns={'index': 'Id'}, inplace=True)
Product['Id'] = Product['Id'].astype(int)
Product.head()
```
